### Let's look at Metabolite Ratios

```r
library(data.table)
```

```r
tmp2 <- TE35_Qfiltered
tcholist <- c()
#naalist <- c()
tcrlist <- c()

for(i in unique(TE35_Qfiltered$Subject)){
    tcho <- TE35_Qfiltered[ which( TE35_Qfiltered$Subject == i & TE35_Qfiltered$Metabolite == "TCho") , ]$Conc    
    naa <- TE35_Qfiltered[ which( TE35_Qfiltered$Subject == i & TE35_Qfiltered$Metabolite == "NAA") , ]$Conc    
    tcr <- TE35_Qfiltered[ which( TE35_Qfiltered$Subject == i & TE35_Qfiltered$Metabolite == "TCr") , ]$Conc    
    tmplist <- TE35_Qfiltered[TE35_Qfiltered$Subject == i, ]$Conc / tcho
    tcholist <- c(tcholist,tmplist)
    tmplist <- TE35_Qfiltered[TE35_Qfiltered$Subject == i, ]$Conc / tcr
    tcrlist <- c(tcrlist,tmplist)
    #tmplist <- TE35_Qfiltered[TE35_Qfiltered$Subject == i, ]$Conc / naa
    #naalist <- c(naalist,tmplist)
}
length(tcholist)
length(tcrlist)
#length(naalist)

tmp2$RatioTCho <- tcholist
tmp2$RatioTCr <- tcrlist
#tmp2$RatioNAA <- naalist
tmp2
```

```r
nrow(TE35_Qfiltered)
```

```r
ggplot(aes(x = Metabolite, y = RatioTCho), data = tmp2) + stat_summary(fun.y = "mean", geom = "bar")
```

```r
ggplot(aes(x = Metabolite, y = RatioTCr), data = tmp2) + stat_summary(fun.y = "mean", geom = "bar")
```
